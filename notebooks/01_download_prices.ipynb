```python
import yfinance as yf
import pandas as pd

tickers = ["AAPL", "MSFT", "GOOGL"]
data = yf.download(
    tickers,
    start="2022-01-01",
    end="2025-01-01",
    threads=False,
    auto_adjust=False,   # ðŸ‘ˆ important
)

adj = data["Adj Close"].reset_index()  # now this works
adj = adj.melt(
    id_vars="Date",
    var_name="ticker",
    value_name="adj_close",
)

adj.to_csv("../data/prices_raw.csv", index=False)
```
